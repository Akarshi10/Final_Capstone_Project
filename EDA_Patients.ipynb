# Cleaning the Data

```python
patientInfo.isnull().sum()
```

```python
patientInfo['age_range_numeric'] = patientInfo.age.str[:2]
patientInfo.head()
```

```python
def stringremover(doc):
    l = doc.split()
    for i in l:
        try:
            num = float(i)
            return num
        except:
            continue
```

```python
contactn=[]
patientInfo['contact_number'] = patientInfo['contact_number'].fillna("0.0")
for i in range(len(patientInfo['contact_number'])):
    doc = patientInfo['contact_number'][i]
    res = stringremover(doc)
    contactn.append(res)
patientInfo['contact_number']=contactn
```

```python
age = []
patientInfo['age_range_numeric'] = patientInfo['age_range_numeric'].fillna("0.0")
for i in range(len(patientInfo['age_range_numeric'])):
    doc = patientInfo['age_range_numeric'][i]
    res = stringremover(doc)
    age.append(res)
patientInfo['age_range_numeric'] = age
```

```python
patientInfo.drop(patientInfo[patientInfo['age'] == "0s"].index, inplace = True)
patientInfo.drop(patientInfo[patientInfo['age_range_numeric'] == 0.0].index, inplace = True)
patientInfo.drop(patientInfo[patientInfo['contact_number'] == 0.0].index, inplace = True)
```
